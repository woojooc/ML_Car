```python
import requests
import time
from bs4 import BeautifulSoup
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager

# 크롬 웹드라이버 설치 및 설정
chrome_options = webdriver.ChromeOptions()

# user 로 보이기위한 user_agent// 각자의 컴퓨터에 맞게 확인 필요
user_agent = 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36'
chrome_options.add_argument('user-agent=' + user_agent)

chrome_options.add_experimental_option('detach', True)
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=chrome_options)
driver.maximize_window()

url = 'https://www.bobaedream.co.kr/dealguide/market_price_new.php'

df_cars=[]

count = -1

# 크롬 열기 및 주소입력
driver.get(url)
driver.implicitly_wait(5)
time.sleep(1)

brand = []
#model = {brand:[]}
model = {}

```

```python
# 매물 선택
driver.find_element(By.XPATH, '//*[@id="frm_search"]/div/div/div[1]/div/dl/dd[1]/button').click()

driver.implicitly_wait(5)
time.sleep(2)

# 페이지 읽기
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
time.sleep(1)
```

```python
# 브랜드명 구역 추출하기
name_sections = soup.find("dl", attrs={"class": "group-list"})
#print(name_sections)
tmp = name_sections.find_all("span", class_="t1")
#print(tmp)
for item in tmp:
    a = item.get_text()
    brand.append(a)
brand
```

```python
#해당 브랜드에 모델명 추가하기
model_section = soup.find_all("dl", attrs={"class": "group-list"})[1]
#print(model_section)
tmp = model_section.find_all("span", class_="t1")
model = {brand[0]:[]}
for item in tmp:
    a = item.get_text()
    model[brand[0]].append(a)
model
```

```python

def crawling(brand):


    # 매물 선택
    driver.find_element(By.XPATH, '//*[@id="frm_search"]/div/div/div[1]/div/dl/dd[1]/button').click()

    driver.implicitly_wait(5)
    time.sleep(2)

    # 페이지 읽기
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    time.sleep(1)

    # 브랜드명 구역 추출하기
    name_sections = soup.find("dl", attrs={"class": "group-list"})
    #print(name_sections)
    tmp = name_sections.find_all("span", class_="t1")
    #print(tmp)
    for item in tmp:
        a = item.get_text()
        brand.append(a)
    #brand
    #print(brand)
    #print(type(brand))
    #len = len(brand)

    for i in range(len(brand)):
        # 브랜드 선택 -> 매물 확인
        path = '//*[@id="frm_search"]/div/div/div[1]/div/dl/dd[{0}]/button'
        path_com = path.format(str(i+1))
        driver.find_element(By.XPATH, path_com).click()

        driver.implicitly_wait(5)
        time.sleep(2)

        # 페이지 읽기
        html = driver.page_source
        soup = BeautifulSoup(html, 'html.parser')
        time.sleep(1)

        #해당 브랜드에 모델명 추가하기
        model_section = soup.find_all("dl", attrs={"class": "group-list"})[1]
        #print(model_section)
        tmp = model_section.find_all("span", class_="t1")
        model[brand[i]]=[]
        for item in tmp:
            a = item.get_text()
            model[brand[i]].append(a)
        #print(model)

    return model
    
brand=[]
model={}

#국내차
model = crawling(brand)


# 매물 선택
driver.find_element(By.XPATH, '//*[@id="used-price"]/div[2]/div[1]/ul/li[2]/button').click()

driver.implicitly_wait(5)
time.sleep(2)

#수입차
brand_ex=[]
model_ex={}
model_ex = crawling(brand)

print(brand)
print(model)
print(brand_ex)
print(model_ex)
```
