```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

################################################################################
# parser해도 본문 내용이 보이지 않아서 셀레니움을 사용해야 할 것으로 추정 됌.
################################################################################


base_url = 'https://www.kbchachacha.com/public/search/main.kbc#!?page={idx}&sort=-orderDate&directYn=Y'

df_cars=[]

# page번호를 파라미터로 받은 url들을 모아두는 변수
urls=[]

# max페이지넘버를 담는 변수 | 23-06-13 기준 직거래 매물은 39페이지 까지 존재한다.
maxPage= 40

for page in range(1, maxPage):
    url= base_url.format(idx=str(page))
    urls.append(url)

cols=["이름","링크",'보험이력',"연식","주행거리","연평균주행거리","연료","배기량","색상","가격","위치",'6개월 후','1년 후','2년 후']
    
#모든 url마다 실행
for url in urls:
    res=requests.get(url)
    res.raise_for_status() # 응답 상태 코드를 확인하는 함수 | 오류 코드일 경우 예외를 발생시킨다. | 따로 예외처리를 하지 않았으므로 기본적인 예외 처리를 한다.
    requests.adapters.DEFAULT_RETRIES = 100000 # 요청 실패시 재시도 횟수를 설정한다. | 큰 값을 할당하여 실행 시간이 늘어날 수 있다.
    soup=BeautifulSoup(res.text,"lxml")
    
    cars=soup.find_all('div', class_='thumnail')

    links=[]
    #한 url마다 들어있는 모든 차들에 대해 실행
    for car in cars:
        link = "https://www.kbchachacha.com" + car.a["href"]
        links.append(link)
    for link in links:
        print(link)

        res2=requests.get(link,timeout=5)
        res2.raise_for_status()
        soup2 = BeautifulSoup(res2.text, "lxml")
        # 차량 정보가 담긴 div
        car_info1 = soup2.find("div", attrs={"class": "detail-info01"})

        div_tag = soup2.find('div', class_='car-buy-price')
        name= div_tag.find('strong')
        
        li_tag = soup2.find('li', class_='used02')
        history = li_tag.find('span')
        
        state=soup2.find("div",attrs={"class":"detail-info-table"})
        div_tag = soup2.find('div', class_='txt-info')
        span_list = div_tag.find_all('span')
        location = span_list[3]
        year=state.find("th",text='연식').find_next_sibling("td")
        km=state.find("th",text='주행거리').find_next_sibling("td")
        fuel=state.find("th",text='연료').find_next_sibling("td")
        amount=state.find("th",text='배기량').find_next_sibling("td")
        color=state.find("th",text='색상').find_next_sibling("td")
        div_tag = soup2.find('div', class_='km-txt')
        km_mean = div_tag.find('strong')
    
        div_tag = soup2.find('div', class_='amchart-chart-div')
        span_list = div_tag.find_all('tspan')
        price= span_list[0]
        after6_price = span_list[1]
        after12_price = span_list[2]
        after24_price = span_list[3]
        
        temp=[name.get_text(),link,year.get_text(),history.get_text(),year.get_text(),km.get_text(),km_mean.get_text(),\
          fuel.get_text(),amount.get_text(),color.get_text(),location, price, after6_price, after12_price, after24_price]
        df_cars.append(temp)
df_cars=pd.DataFrame(data=df_cars,columns=cols)
df_cars.to_csv('kb_cars.csv')
print(df_cars.head())
```
